```python
charsTest = pd.read_csv('emnist-byclass-test.csv', header=None)
```

```python
charsTest.sample(frac = 1)
```

```python
y_train = np.ravel(pd.DataFrame(charsTrain.iloc[:30000, 0]))
y_test = np.ravel(pd.DataFrame(charsTest.iloc[:30000, 0]))
X_train = pd.DataFrame(charsTrain.iloc[:30000, 1:])
X_test = charsTest.drop(charsTest.iloc[:, 0])
```

```python
X_train.shape
```

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
params = {'kernel':('linear', 'rbf'), 'C':[1, 10, 100]}
```

```python
classes = []
for i in range(62):
    i += 1
    train = (y_train == i)
    test = (y_test == i)
    svc = svm.SVC(gamma="scale", C= 1, kernel = "linear")
    classes.append([train, test, svc])
```

```python
svc.fit(X_train, classes[0][0])
```
